### Scrapping the other 2 missing documents

```python
import fitz
```

```python
def extract_document(pdf_filename, output_filename):
    document = fitz.open(pdf_filename)

    print(f"File {pdf_filename} has {document.page_count} pages.")

    complete_doc = ""

    for page in range(document.page_count):
        complete_doc += document.load_page(page).get_text('text')

    with open(output_filename, "w") as output_file:
        output_file.write(complete_doc)

    return complete_doc
```
