```python
all_recents = []

for question in questions:
    refs = list(question['all_formatted_references'].keys())

    for reference in refs:
        name_parts = split_document_name(reference.split(" - ")[0])
    
        try:
            doc_year = int(name_parts[-1].split()[-1])
    
            if doc_year >= 2023:
                all_recents.append({"question": question['question_number'], 
                                    "reference": reference})
        except:
            print(f">>{reference}")
```
